**CANCER DE MAMA EM WISCONSIN**

```
!pip install ucimlrepo

from sklearn import tree
from sklearn.model_selection import train_test_split
from sklearn import metrics

from ucimlrepo import fetch_ucirepo
import pandas as pd

BREAST_CANCER_WISCONSIN_DIAGNOSTIC = 17
SPAM = 94

# Dados para cancer de mama em Wisconsin
data_set_heart_disease = fetch_ucirepo(id=BREAST_CANCER_WISCONSIN_DIAGNOSTIC)
y_heart_disease = data_set_heart_disease.data.targets
X_heart_disease = data_set_heart_disease.data.features

# Dados para classificação de e-mail (spam ou não-spam)
data_set_spam = fetch_ucirepo(id=SPAM)
y_spam = data_set_spam.data.targets
X_spam = data_set_spam.data.features

#url = "https://raw.githubusercontent.com/tmoura/machinelearning/master/datasets/iris.data"

# Carregar base de dados
# DataFrame
# dataset = pd.read_csv(url, header=None)

# columns = len(dataset.columns)

# y = dataset[0] # extrai a primeira coluna, que é o label
# X = dataset.loc[:,1:columns-1]
```

```
X_train, X_test, y_train, y_test = train_test_split(X_heart_disease, y_heart_disease, test_size=0.2, random_state=None, stratify=y_heart_disease) # 80% treino e 20% teste
```

```
# Treinamendo da Árvore de Decisão
#model = tree.DecisionTreeClassifier(criterion="entropy",max_depth=2)
model = tree.DecisionTreeClassifier(criterion="entropy")
model = model.fit(X_train, y_train)
```

```
# Predição e Resultados

result = model.predict(X_test)
acc = metrics.accuracy_score(result, y_test)
show = round(acc * 100)
print("{}%".format(show))

print(list(result))
print(list(y_test))
```

```
tree.plot_tree(model,filled=True)
```
